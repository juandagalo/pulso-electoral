## Import libraries

```python
from __future__ import annotations

import json
import sys
from pathlib import Path

import pandas as pd

sys.path.insert(0, str(Path("../../src").resolve()))

# NOTE: BERTopic was removed. Use compute_embeddings for embeddings,
# then sklearn for clustering (see Topic Extraction cell below).
from nlp.topics import compute_embeddings
```

## Load data

```python
with open("../../data/sample/sample_posts.json") as f:
    posts = json.load(f)
texts = [p["text"] for p in posts]
print(f"Loaded {len(texts)} texts for topic modeling")
```

## Topic Extraction

```python
# NOTE: BERTopic was removed. Topic modeling now uses
# sentence-transformers embeddings + sklearn clustering directly.
# Full pipeline: compute_embeddings -> sklearn.cluster -> label topics

try:
    from sklearn.cluster import KMeans

    embeddings = compute_embeddings(texts)
    n_clusters = min(5, len(texts))  # Adjust based on corpus size
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(embeddings)
    print(f"Discovered {len(set(labels))} topic clusters")
    print(f"Cluster sizes: {pd.Series(labels).value_counts().to_dict()}")
except Exception as e:
    print(f"Topic modeling requires more data: {e}")
```

## Analysis of Results and Conclusions

Topic modeling reveals dominant narrative clusters. With full data,
expected topics: elections, corruption, civic space, digital manipulation,
peace process, economic concerns.
