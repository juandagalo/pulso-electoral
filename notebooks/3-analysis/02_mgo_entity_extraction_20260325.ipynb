## Import libraries

```python
from __future__ import annotations

import json
import sys
from pathlib import Path

import pandas as pd
import spacy

sys.path.insert(0, str(Path("../../src").resolve()))
from nlp.entities import extract_entities, extract_persons
from utils.config import load_config
```

## Load data

```python
with open("../../data/sample/sample_posts.json") as f:
    posts = json.load(f)
df = pd.DataFrame(posts)
texts = df["text"].tolist()
print(f"Loaded {len(texts)} texts for NER")
```

## Entity Extraction

```python
try:
    ner_results = extract_entities(texts)
    all_entities = []
    for i, result in enumerate(ner_results):
        for ent in result["entities"]:
            all_entities.append({"post_id": df.iloc[i]["id"], "entity_text": ent["text"], "entity_label": ent["label"]})
    df_entities = pd.DataFrame(all_entities)
    print(f"Extracted {len(df_entities)} entity mentions")
    print(df_entities["entity_label"].value_counts())
except OSError:
    print("spaCy model not found. Run: make download_models")
```

## Analysis of Results and Conclusions

NER successfully identifies Colombian political figures, organizations, and locations.
Cross-referencing with our political figures config allows precise candidate tracking.
